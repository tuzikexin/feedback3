```python
codeshuwen = ''
```

```python
import base64
from tqdm.notebook import tqdm
import os

shuwen = base64.b64decode(base64.b64decode(base64.b64decode(codeshuwen))).decode('UTF-8')
```

```python
shuwenls = shuwen.split('Freqyoryoiqq')
```

```python
for i in range(len(shuwenls)):
    with open('./shuwendaima/shuwen{}.py'.format(i), 'w') as f:
        f.write(shuwenls[i])
    f.close()
```

```python
for i in tqdm(range(len(shuwenls))):
    os.system('python ./shuwendaima/shuwen{}.py'.format(i))
```

```python
for i in range(1, 15):
    shuwenfile = '/kaggle/input/shuwen-svt/shuwen_{}.py'.format(i)
    with open(shuwenfile, 'r') as f:
        shuwen = f.read()
    f.close()
    code = base64.b64decode(base64.b64decode(base64.b64decode(shuwen))).decode('UTF-8')
    with open('./shuwensvt/{}.py'.format(i), 'w') as f:
        f.write(code)
    f.close()
```

```python
for i in tqdm(range(1, 13)):
    os.system('python ./shuwensvt/{}.py'.format(i))
```
